# Data Ingestion & Database Setup

Objective:
- Load raw mutual fund datasets
- Validate structure
- Store data in SQLite database
- Verify successful ingestion


```python
import pandas as pd
import sqlite3
import os
from pathlib import Path
```

```python
raw_path = "../data/raw"

files = os.listdir(raw_path)

print(f"Total files found: {len(files)}")

for f in files:
    print(f)
```

```python
fund_perf = pd.read_csv("../data/raw/07_scheme_performance.csv")
invest_txn = pd.read_csv("../data/raw/08_investor_transactions.csv")
portfolio = pd.read_csv("../data/raw/09_portfolio_holdings.csv")

print(fund_perf.shape)
print(invest_txn.shape)
print(portfolio.shape)
```

```python
datasets = {
    "scheme_performance": fund_perf,
    "investor_transactions": invest_txn,
    "portfolio_holdings": portfolio
}

for name, df in datasets.items():
    print("\n", "="*40)
    print(name)
    print("Rows, Columns:", df.shape)
    print("Missing Values:")
    print(df.isnull().sum().sum())
    print("Duplicates:", df.duplicated().sum())
```

```python
conn = sqlite3.connect("../data/db/bluestock_mf.db")

tables = pd.read_sql(
    """
    SELECT name
    FROM sqlite_master
    WHERE type='table'
    """,
    conn
)

tables
```
